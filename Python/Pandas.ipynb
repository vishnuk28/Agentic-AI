```python
def Grade(Score):
  if Score >=90:
    return 'A'
  elif Score >=80:
    return 'B'
  elif Score >=70:
    return 'C'
  elif Score >=60:
    return 'D'
  else:
    return 'F'
#apply() function
#adding new columns by using the logic or conditions
df['Grade_assigned'] = df['Score'].apply(Grade)
print(df)
```

```python
print(df)
```

```python
result={
    True:"Pass",
    False:"Fail"
}
Passs={'A':90,"B":80,"D":70}
df['pass']=df['T|F'].map(Passs)
print(df)
```

```python
df['result']=df['T|F'].map(result)
print(df)
```

```python
df.drop('pass',axis=1, inplace=True)
print(df)
```

```python
df.rename(columns={'result':'Result','Name':'Student'},inplace=True)
print(df)
```

```python
df.replace('b',np.nan,inplace=True)
print(df)
```
